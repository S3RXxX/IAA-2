```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
import warnings
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTENC

InteractiveShell.ast_node_interactivity = "all"
pd.set_option('display.precision', 3)
sns.set()
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("./raw_data.csv")
```

```python
X = df.drop(columns=['Y'])
Y = df['Y']
```

Dividimos el dataset

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state = 10)
```

Normalizamos las variables númericas

```python
from sklearn import preprocessing

min_max_scaler =  preprocessing.MinMaxScaler()
normalizated_train = X_train.copy()

normalizated_train[:7] = min_max_scaler.fit_transform(X_train[:7])

normalizated_test = X_test.copy()
normalizated_test[:7] = min_max_scaler.transform(X_test[:7])
```

```python
smote = SMOTENC(random_state=42,  categorical_features=list(range(7,len(X_train.columns))))
```

```python
Y_train.value_counts()
```

Hacemos oversampling de la variable objetivo

```python
X_train_resampled, Y_train_resampled = smote.fit_resample(normalizated_train, Y_train)
```

```python
Y_train_resampled.value_counts()
Y_test.value_counts()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, recall_score, f1_score
from sklearn.model_selection import GridSearchCV
```

Definimos con que parámetros experimentaremos

```python
search_space = {
    'n_neighbors':range(2, 50),
    'metric': ['correlation', 'euclidean', 'cosine', 'minkowski', 'cityblock', 'jaccard', 'gower']
}
```

```python
knn = KNeighborsClassifier()
```

Definimos que métricas utlizaremos para evaluear el modelo y preparamos el modelo para hacer cross-validation

```python
scoring = ['f1', 'recall', 'accuracy']
refit = 'recall'
GS = GridSearchCV(estimator=knn,
                 param_grid=search_space,
                 scoring=scoring,
                 refit = refit,
                 cv = 5,
                 verbose = 4)
```

Entrenamos al modelo

```python
_ = GS.fit(X_train_resampled, Y_train_resampled)
```

```python
print(GS.best_estimator_)
```

```python
print(GS.best_params_)
```

```python
print(GS.best_score_)
```

Ordenamos el resultado y lo guardamos en un csv

```python
result = pd.DataFrame(GS.cv_results_)
result = result.sort_values("mean_test_f1", ascending=False)
result.to_csv('cv_results_knn.csv')
result
```

Hacemos un gráfico para visualizar el rendimiento del modelo dependiendo de los hyperparametros

```python
import seaborn as sns

sns.lineplot(data=result, x="param_n_neighbors", y="mean_test_f1", hue="param_metric")
plt.legend(loc='lower right')
```

```python
best_knn = KNeighborsClassifier(n_neighbors=10, metric='jaccard')
best_knn.fit(X_train_resampled, Y_train_resampled)
```

```python

y_preds  = best_knn.predict(normalizated_test)
accuracy_score(Y_test, y_preds)
recall_score(Y_test, y_preds)
f1_score(Y_test, y_preds)



print(classification_report(Y_test, y_preds))
```

```python
matrix = confusion_matrix(Y_test.tolist(),  
                                            y_preds.tolist())#turn this into a dataframe

labels = ['stayed', 'churned']
matrix_df = pd.DataFrame(matrix)#plot the result
ax = plt.axes()
_ = sns.set(font_scale=1.3)
_ = plt.figure(figsize=(10,7))
_ = sns.heatmap(matrix_df, annot=True, fmt="g", ax=ax, cmap="magma")#set axis titles
_ = ax.set_title('Confusion Matrix')
_ = ax.set_xlabel("Predicted label", fontsize =15)
_ = ax.set_xticklabels(labels)
_ = ax.set_ylabel("True Label", fontsize=15)
_ = ax.set_yticklabels(labels, rotation = 0)
_ = plt.show()
```
