## Partició

```python
def seed_finder():
    seed = 0
    while True:
        for seed2 in range(seed+1):
            train, test_val = train_test_split(Ciutats, test_size=0.2, random_state=seed)
            validation, test = train_test_split(test_val, test_size=0.5, random_state=seed2)
            if len(train["State"].value_counts())==17 and len(validation["State"].value_counts())==17 and len(test["State"].value_counts())==17:
                return seed, seed2
        seed += 1
# print(seed_finder())
```

```python
# minmax scale Male Population
min_max_scaler =  preprocessing.MinMaxScaler()
min_max_scaler.fit(Train_norm[['Male Population']])
Train_norm[['Male Population']]=min_max_scaler.transform(Train_norm[['Male Population']])
Val_norm[['Male Population']]=min_max_scaler.transform(Val_norm[['Male Population']])
Test_norm[['Male Population']]=min_max_scaler.transform(Test_norm[['Male Population']])
```

```python
fig, axes= plt.subplots(1,2, gridspec_kw={'width_ratios': [1, 4]}, figsize=(9,5))
Train_norm.boxplot(column='Male Population',ax=axes[0]);
Train_norm.hist(column='Male Population', ax=axes[1]);
```

```python
# minmax scale Female Population
min_max_scaler =  preprocessing.MinMaxScaler()
min_max_scaler.fit(Train_norm[['Female Population']])
Train_norm[['Female Population']]=min_max_scaler.transform(Train_norm[['Female Population']])
Val_norm[['Female Population']]=min_max_scaler.transform(Val_norm[['Female Population']])
Test_norm[['Female Population']]=min_max_scaler.transform(Test_norm[['Female Population']])
```

```python
#norm boxcox
x, lb= boxcox(Train['Total Population'])
Train_norm['Total Population'] = x

y = boxcox(Val['Total Population'], lb)
Val_norm['Total Population'] = y

z = boxcox(Test['Total Population'], lb)
Test_norm['Total Population'] = z

# minmax scale
min_max_scaler =  preprocessing.MinMaxScaler()
min_max_scaler.fit(Train_norm[['Total Population']])
Train_norm[['Total Population']]=min_max_scaler.transform(Train_norm[['Total Population']])
Val_norm[['Total Population']]=min_max_scaler.transform(Val_norm[['Total Population']])
Test_norm[['Total Population']]=min_max_scaler.transform(Test_norm[['Total Population']])
```

```python
#log
Train_norm['Number of Veterans']=Train_norm['Number of Veterans'].apply(log)
Val_norm['Number of Veterans']=Val_norm['Number of Veterans'].apply(log)
Test_norm['Number of Veterans']=Test_norm['Number of Veterans'].apply(log)

# minmax scale Number of Veterans
min_max_scaler =  preprocessing.MinMaxScaler()
min_max_scaler.fit(Train_norm[['Number of Veterans']])
Train_norm[['Number of Veterans']]=min_max_scaler.transform(Train_norm[['Number of Veterans']])
Val_norm[['Number of Veterans']]=min_max_scaler.transform(Val_norm[['Number of Veterans']])
Test_norm[['Number of Veterans']]=min_max_scaler.transform(Test_norm[['Number of Veterans']])
```

```python
# minmax scale Foreign-born
min_max_scaler =  preprocessing.MinMaxScaler()
min_max_scaler.fit(Train_norm[['Foreign-born']])
Train_norm[['Foreign-born']]=min_max_scaler.transform(Train_norm[['Foreign-born']])
Val_norm[['Foreign-born']]=min_max_scaler.transform(Val_norm[['Foreign-born']])
Test_norm[['Foreign-born']]=min_max_scaler.transform(Test_norm[['Foreign-born']])
```

```python
#norm boxcox
x, lb= boxcox(Train['Average Household Size'])
Train_norm['Average Household Size'] = x

y = boxcox(Val['Average Household Size'], lb)
Val_norm['Average Household Size'] = y

z = boxcox(Test['Average Household Size'], lb)
Test_norm['Average Household Size'] = z


# minmax scale
min_max_scaler =  preprocessing.MinMaxScaler()
min_max_scaler.fit(Train_norm[['Average Household Size']])
Train_norm[['Average Household Size']]=min_max_scaler.transform(Train_norm[['Average Household Size']])
Val_norm[['Average Household Size']]=min_max_scaler.transform(Val_norm[['Average Household Size']])
Test_norm[['Average Household Size']]=min_max_scaler.transform(Test_norm[['Average Household Size']])
```

```python
#min max Median Age
min_max_scaler =  preprocessing.MinMaxScaler()
min_max_scaler.fit(Train_norm[['Median Age']])
Train_norm[['Median Age']]=min_max_scaler.transform(Train_norm[['Median Age']])
Val_norm[['Median Age']]=min_max_scaler.transform(Val_norm[['Median Age']])
Test_norm[['Median Age']]=min_max_scaler.transform(Test_norm[['Median Age']])
```
